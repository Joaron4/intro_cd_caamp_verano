<h2 align="left" style="font-size:3vw">
  <span style="color:#00c389">1. Operadores de Comparación</span>
</h2>


* Expresiones Booleanas hacen una pregunta que produce una respuesta de sí o no que utilizamos para controlar el flujo del programa
* Las expresiones Booleanas usando operadores de comparación evalúan True / False o Yes / No
* Estos operadores se fijan en las variables pero no cambian las variables


|     Python    |       Significado      |
|:-------------:|:----------------------:|
|        <      |        Menor que       |
|       <=      |     Menor o igual a    |
|       ==      |        Igual que       |
|       >=      |     Mayor o igual a    |
|        >      |        Mayor que       |
|       !=      |      No es igual a     |


Recuerden que `“=”` se usa para asignación de variables

````python
x = 5
if x == 5 : 
    print('Es igual a 5')
if x > 4 : 
    print('Mayor que 4')
if  x >= 5 :
    print('Mayor o igual a 5')
if x < 6 : print('Menor que 6') 
if x <= 5 :
    print('Menor o igual a 5')
if x != 6 :
    print('No es igual a 6')
````

Para esto utilizaremos la palabra reservada:

```python

```

````python
x = 7
print('Antes de 5')
if  x == 5 :
    print('Es 5')
    print('Aún 5')
    print('Tercer 5')
print('Después de 5')
print('Antes de 6')
if x == 6 :
    print('Es 6')
    print('Aún es 6')
    print('Tercer 6')
print('Después de 6')
````

<h2 align=left style="font-size:3vw"><span style="color:#00c389">3. Indentación:</span></h2>

Son 4 espacios en blanco (no te preocupes jupyter notebook lo pone de manera automática). 

Debes tener en cuenta que:

* Aumentar indentación después de un if o un for (después del : )
* Mantener indentación para indicar el alcance del bloque (qué líneas se ven afectadas por el if/for)
* Reducir indentación al nivel del if o el for para indicar el fin del bloque
* Líneas vacías son ignoradas – no afectan la indentación
* Comentarios en una línea por sí solos son ignorados en lo que tiene que ver con indentación

### Aumentar / mantener luego de if o for
### Reducir para indicar el fin del bloque



````python
x = 5
if x > 2 :
    print('Mayor que 2')
    print('Aún mayor')
print('Acabamos con el 2')

 
````

````python
for i in range(5):
    print(i)
    if i > 2 : 
        print('Mayor que 2')
    print('Acabamos con i', i)
print('Terminado')
````

<h2 align=left style="font-size:3vw"><span style="color:#00c389">5. Decisiones de doble vía:</span></h2>


* En ocasiones queremos hacer una cosa si alguna expresión condicional es verdadera y otra cosa diferente si es falsa
* Es como una Y – debemos escoger un camino o el otro, pero no ambos


La palabra reservada:

```python
else

```

Nos va a ser útil para crear esta doble vía

````python
x = 2

if x > 2 :
    print('Mayor')
else:
    print('Menor')
    print('Menor otra vez')

print('Todo Listo!')
````

Piensen en else como una malla de seguridad.

<h1 align=center style="font-size:3vw"><span style="color:#ff585d">Try/Except</span></h1>


* Se rodea una sección peligrosa del código con `try` y `except`
* Si el código dentro del `try` funciona, el `except` se salta
* Si el código en el `try` falla, se salta a la sección de `except`


````python
astr = 'Hola Paola'
istr = int(astr)
print('Primero', istr)
astr = '123'
istr = int(astr)
print('Segundo', istr)
````

````python
astr
````

````python
astr = 'Hola Paola'
try:
    istr = int(astr)
except:
    istr = -1

print('Primero', istr) #Cuando la primer conversión falla, se va para la cláusula except: y el programa continúa.

astr = '123'
try:
    istr = int(astr)
except:
    istr = -1

print('Segundo', istr)  #Cuando la segunda conversión es exitosa, - se salta la cláusula except: y el programa continúa.
````

````python
astr = 'Paola'
try:
    print('Hola') 
    istr = int(astr)
    print('Tú') 
except:
    istr = -1

print('Hecho', istr) 
````

Para Terminar, ejecute la celda de abajo `ctrl+enter`, ingrese su **Apellido** y su **Nombre**. En es orden. Luego, haga clic en "Ingresar Nombre". 🔻🔻🔻🔻🔻🔻🔻

````python
from ipywidgets import interact, widgets
from IPython.display import display, Javascript, HTML
import unicodedata

def remove_accents(input_str):
    nfkd_form = unicodedata.normalize('NFKD', input_str)
    return ''.join([char for char in nfkd_form if not unicodedata.combining(char)])

text = widgets.Text(
    value='',
    placeholder='Escriba su apellido y nombre acá',
    description='Nombre:',
    disabled=False,
    style={'description_width': 'initial', 'font_weight': 'bold', 'width': '400px'}  # Adjust the width here
)

button = widgets.Button(description="Ingresar Nombre")

output = widgets.Output()

saved_value = ""

def callback(change):
    global saved_value
    normalized_text = remove_accents(change.new)
    saved_value = normalized_text.lower().strip().replace(' ', '_').replace('  ', '_')
    

def on_button_click(button):
    with output:
        js_code = f"Jupyter.notebook.rename('{saved_value}_taller_2');"
        display(Javascript(js_code))
        
        


text.observe(callback, names='value')
button.on_click(on_button_click)

# Use HTML to create a toggle button for code cells
html_toggle_code = HTML('''
    <script>
    code_show=true; 
    function code_toggle() {
        var cell = Jupyter.notebook.get_cell(-2);
        cell.element.find('.input').hide();
        
    } 
    $( document ).ready(code_toggle);
    </script>
    
''')

# Display the HTML toggle button along with the widget and output
display(html_toggle_code)
display(text)
display(button)
display(output)
````
