<h2 align=left style="font-size:3vw"><span style="color:#00c389">2. ¿Qué es una función?</span></h2>





* En Python una función es código reutilizable que recibe argumentos como entrada, hace algunos procesos computacionales, y luego retorna el resultado o resultados.
* Definimos una función usando la palabra reservada **`def`**.
* Invocamos la función al usar el nombre de la función, paréntesis, y argumentos en una expresión.

<h2 align=left style="font-size:3vw"><span style="color:#00c389">3. Funciones <em>built-in</em> :</span></h2>


Las funciones "built-in" de Python son funciones integradas que vienen predefinidas en el lenguaje y están siempre disponibles sin necesidad de importarlas. 

Ejemplos comunes incluyen `print()`, `len()`, `range()`, `type()`, y `int()`. Estas funciones permiten realizar operaciones básicas y comunes, como mostrar texto en la pantalla, obtener la longitud de una lista, generar secuencias de números, determinar el tipo de una variable, y convertir datos a otros tipos, facilitando la programación y el desarrollo de aplicaciones.

<h2 align=left style="font-size:3vw"><span style="color:#00c389">7. Parámetros:</span></h2>

Un parámetro es una variable que usamos en la definición de la función.  Permite al código en la función acceder a los argumentos para una invocación específica de la función.

````python
def saludo(lang):
    if lang == 'es':
        print('Hola')
    elif lang == 'fr':
        print('Bonjour')
    else:
        print('Hello')

        
saludo('en')
````

<h2 align=left style="font-size:3vw"><span style="color:#00c389">8. Valor de Retorno (return):</span></h2>



A menudo una función recibe sus argumentos, hace algún proceso computacional, y retorna un valor para ser usado como el valor de la llamada de la función en la expresión de llamada.  La palabra `return` se usa para esto.

````python
def saludo():
    return "Hello"

print(saludo(), "Marco")
print(saludo(), "Gladys")
````

<h2 align=left style="font-size:3vw"><span style="color:#00c389">9. Función fructífera:</span></h2>

<h2 align=left style="font-size:3vw"><span style="color:#00c389">10. Argumentos, Parámetros, y Resultados:</span></h2>



* Podemos definir más de un parámetro en la definición de la función
* Simplemente habría que incluir más argumentos cuando llamemos la función
* El número y el orden de los parámetros y los argumentos deben coincidir


````python
def sumados(a,b):
    added = a + b
    return added

x = sumados(5,3)
print(x)
````

````python
def sumados(alto, ancho):
    added = alto - ancho
    return added

x = sumados(3,5)
print(x)
````

````python
sumados(4,15)
````

````python
added = 4 + 5
added
````

<h2 align=left style="font-size:3vw"><span style="color:#00c389">11. Funciones Vacías (non-fruitful):</span></h2>


* Cuando una función no retorna un valor, le decimos una función “vacía”
* Las funciones que retornan valores son fructíferas (fruitful)
* Las funciones vacías son “no fructíferas”

Para Terminar, ejecute la celda de abajo `ctrl+enter`, ingrese su **Apellido** y su **Nombre**. En es orden. Luego, haga clic en "Ingresar Nombre". 🔻🔻🔻🔻🔻🔻🔻

````python
from ipywidgets import interact, widgets
from IPython.display import display, Javascript, HTML
import unicodedata

def remove_accents(input_str):
    nfkd_form = unicodedata.normalize('NFKD', input_str)
    return ''.join([char for char in nfkd_form if not unicodedata.combining(char)])

text = widgets.Text(
    value='',
    placeholder='Escriba su apellido y nombre acá',
    description='Nombre:',
    disabled=False,
    style={'description_width': 'initial', 'font_weight': 'bold', 'width': '400px'}  # Adjust the width here
)

button = widgets.Button(description="Ingresar Nombre")

output = widgets.Output()

saved_value = ""

def callback(change):
    global saved_value
    normalized_text = remove_accents(change.new)
    saved_value = normalized_text.lower().strip().replace(' ', '_').replace('  ', '_')
    

def on_button_click(button):
    with output:
        js_code = f"Jupyter.notebook.rename('{saved_value}_taller_3');"
        display(Javascript(js_code))
        
        


text.observe(callback, names='value')
button.on_click(on_button_click)

# Use HTML to create a toggle button for code cells
html_toggle_code = HTML('''
    <script>
    code_show=true; 
    function code_toggle() {
        var cell = Jupyter.notebook.get_cell(-2);
        cell.element.find('.input').hide();
        
    } 
    $( document ).ready(code_toggle);
    </script>
    
''')

# Display the HTML toggle button along with the widget and output
display(html_toggle_code)
display(text)
display(button)
display(output)
````
