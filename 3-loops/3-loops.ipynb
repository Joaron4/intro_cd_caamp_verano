<h2 align=left style="font-size:3vw"><span style="color:#00c389">6. Loops definidos:</span></h2>

## Un Loop Definido Sencillo

<h3 align=left style="font-size:3vw"><span style="color:#00a3ad">D. Contando:</span></h3>


````python
zork = 0
print('Antes', zork)
for cosa in [9, 41, 12, 3, 74, 15] :
    zork = zork + 1
    print(zork, cosa)
print('Después', zork)
````

````python
zork = 0
print('Antes', zork)
for cosa in [9, 41, 12, 3, 74, 15] :
    zork += 1
    print(zork, cosa)
print('Después', zork)
````

````python
zork = 0
print('Antes', zork)
for cosa in [9, 41, 12, 3, 74, 15] :
    zork ++
    print(zork, cosa)
print('Después', zork)
````

Para contar cuántas veces ejecutamos un loop, introducimos una variable de conteo que inicia en 0 y añadimos uno a ella cada vez a través del loop.

<h3 align=left style="font-size:3vw"><span style="color:#00a3ad">E. Sumando:</span></h3>

````python
zork = 0
print('Antes', zork)
for cosa in [9, 41, 12, 3, 74, 15] :
    zork = zork + cosa
    print(zork, cosa)
print('Después', zork)
````

Para sumar un valor que encontramos en un loop, introducimos una variable de suma que comienza en 0 y sumamos el valor a la suma cada vez a través del loop.

<h3 align=left style="font-size:3vw"><span style="color:#00a3ad">F. Hallando el promedio:</span></h3>



````python
count = 0
suma = 0
print('Antes', count, suma)
for value in [9, 41, 12, 3, 74, 15] :
    count = count + 1
    suma = suma + value
    print(count, suma, value)
print('Después', count, suma, round(suma/count,2))
````

Un promedio simplemente combina los patrones de conteo y suma y divide cuando el loop termina.

<h3 align=left style="font-size:3vw"><span style="color:#00a3ad">G. Filtrando:</span></h3>


````python
print('Antes')
for valor in [9, 41, 12, 3, 74, 15] :
    if valor > 20:
        print('Número grande',valor)
print('Después')
````

Usamos la orden `if` en el loop para atrapar / filtrar los valores que estamos buscando.

<h3 align=left style="font-size:3vw"><span style="color:#00a3ad">H. Búsqueda usando una variable booleana:</span></h3>


````python
found = False
print('Antes', found)
for valor in [9, 41, 12, 3, 74, 15] : 
    if valor == 3 :
        found = True
    print(found, valor)
print('Después', found)
````

Si solo queremos buscar y conocer si un valor es encontrado, usamos una variable que empieza en `False` y se establece como `True` tan pronto se encuentre lo buscado.

<h3 align=left style="font-size:3vw"><span style="color:#00a3ad">I. Hallando el menor valor:</span></h3>


````python
#Aquí el código para el mayor valor

mayor_por_ahora = -1
print('Antes', mayor_por_ahora)
for the_num in [9, 41, 12, 3, 74, 15] :
    if the_num < mayor_por_ahora :
        mayor_por_ahora = the_num
    print(mayor_por_ahora, the_num)

print('Después', mayor_por_ahora)
````

````python
#Aquí el código para el menor valor

mayor_por_ahora = 100
print('Antes', mayor_por_ahora)
for the_num in [9, 41, 12, 3, 74, 15] :
    if the_num < mayor_por_ahora :
        mayor_por_ahora = the_num
    print(mayor_por_ahora, the_num)

print('Después', mayor_por_ahora)
````

¿Cómo lo cambiamos para que encuentre el menor valor?

````python
menor_por_ahora = -1
print('Antes', menor_por_ahora)
for the_num in [9, 41, 12, 3, 74, 15] :
    if the_num < menor_por_ahora :
        menor_por_ahora = the_num
    print(menor_por_ahora, the_num)

print('Después', menor_por_ahora)
````

¿Qué sucedió?

````python
elmenor = None
print('Antes')
for value in [9, 41, 12, 3, 74, 15] :
    if elmenor is None : 
        elmenor = value
    elif value < elmenor : 
        elmenor = value
    print(elmenor, value)
print('Después', elmenor)
````

Ahora tenemos una variable que representa elmenor número hasta ahora. En el primer recorrido elmenor es None, así que se toma el primer valor como elmenor.
